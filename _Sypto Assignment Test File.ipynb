```python
# Define common column names
columns = ['Instrument Name', 'Symbol', 'Expiry Date', 'Instrument Type']
```

```python
angel= pd.read_csv(r'C:\Users\umade\Downloads\Brokers_data\angel_script_master.csv',usecols=columns)
fyers=pd.read_csv(r'C:\Users\umade\Downloads\Brokers_data\fyers_script_master.csv',usecols=columns)
zerodha=pd.read_csv(r'C:\Users\umade\Downloads\Brokers_data\zerodha_script_master.csv',usecols=columns)
alice=pd.read_csv(r'C:\Users\umade\Downloads\Brokers_data\Alice_script_master.csv',usecols=columns)
```

```python
# Rename columns and add 'Broker' column
zerodha.columns = columns
fyers.columns = columns
angel.columns = columns
alice.columns=columns
```

```python
zerodha['Broker'] = 'Zerodha'
fyers['Broker'] = 'Fyers'
angel['Broker'] = 'Angel'
alice['Broker']='Alice'
```

```python
# Combine data from all brokers
all_data = pd.concat([zerodha, fyers, angel, alice])
```

```python
# Pivot table to map symbols across brokers
mapped_data = all_data.pivot_table(
    index=['Instrument Name', 'Expiry Date', 'Instrument Type'],
    columns='Broker',
    values='Symbol',
    aggfunc='first'
).reset_index()
```

```python
# Rename columns
mapped_data.columns.name = None
mapped_data = mapped_data.rename(columns={
    'Zerodha': 'Zerodha Symbol', 
    'Fyers': 'Fyers Symbol', 
    'Angel': 'Angel Symbol',
    'Alice': 'Alice Symbol'
})
```
